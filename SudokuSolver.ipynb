```python
import cv2
import pytesseract
import numpy as np
import matplotlib.pyplot as plt

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Load the image
image = cv2.imread(r"C:\Users\username\Downloads\sudoku.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, binary = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY_INV)
kernel = np.ones((3, 3), np.uint8)
cleaned_image = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)

# Assuming the grid is 450x450
grid_size = image.shape[0]
cell_size = grid_size // 9

# Split the image into cells and recognize digits
sudoku = []
for i in range(9):
    row = []
    for j in range(9):
        x, y = j * cell_size, i * cell_size
        cell = image[y:y+cell_size, x:x+cell_size]
        cropped_cell = cell[7:-7, 7:-7]  # Crop a small margin
        cell_resized = cv2.resize(cropped_cell, (100, 100))

        custom_config = r'--oem 3 --psm 10 -c tessedit_char_whitelist=0123456789'
        text = pytesseract.image_to_string(cell_resized, config=custom_config)
        digit = text.strip()
        row.append(int(digit) if digit.isdigit() else 0)
    sudoku.append(row)

# Print the recognized Sudoku grid
print("\nRecognized Sudoku Grid:")
for row in sudoku:
    print(row)
# Visualize the cleaned image
plt.figure(figsize=(8, 8))
plt.imshow(cleaned_image, cmap='gray')
plt.title('Cleaned Image')
plt.axis('off')
plt.show()
# Function to update the grid interactively
def modify_grid(sudoku):
    while True:
        print("\nCurrent Grid:")
        print("     " + "  ".join(str(i) for i in range(9)))  # Print column indices
        print("   " + "-" * 25)  # Horizontal separator
        
        for i, row in enumerate(sudoku):
            row_str = "  ".join(str(num) if num != 0 else "." for num in row)  # Replace 0 with "."
            print(f" {i} | {row_str}")  # Print row index and row content
        
        choice = input("\nDo you want to change a cell value? (y/n): ").strip().lower()
        if choice != 'y':
            break
        
        try:
            # Ask the user for the cell to modify
            user_input = input("Enter row, column, and value (e.g., '8,8,8' or '8 8 8'): ").strip()
            
            # Parse the input (supporting formats like `8,8,8`, `8 8 8`, or `8, 8, 8`)
            parts = [int(x) for x in user_input.replace(",", " ").split()]
            if len(parts) != 3:
                print("Invalid input format. Please provide three numbers separated by spaces or commas.")
                continue
            
            row, col, new_value = parts
            if not (0 <= row < 9 and 0 <= col < 9 and 0 <= new_value <= 9):
                print("Invalid row, column, or value. Please enter numbers between 0-8 for row/column and 0-9 for value.")
                continue
            
            # Update the cell value
            print(f"Current value at cell ({row},{col}): {sudoku[row][col]}")
            sudoku[row][col] = new_value
            print(f"Updated cell ({row},{col}) to {new_value}.")
        
        except ValueError:
            print("Invalid input. Please enter numeric values.")
    
    return sudoku



# Modify the grid interactively
sudoku = modify_grid(sudoku)

# Print the final corrected Sudoku grid
print("\nFinal Corrected Sudoku Grid:")
for row in sudoku:
    print(row)
    
graph = {}


# Print the matrix sudokuay
def printMatrix():
    for i in range(0, 9):
        for j in range(0, 9):
            print(str(sudoku[i][j]), end=" ")
        print()


# Method to check if the inserted element is safe
def is_safe(x, y):
    key = sudoku[x][y]
    for i in range(0, 9):
        if i != y and sudoku[x][i] == key:
            return False
        if i != x and sudoku[i][y] == key:
            return False

    r_start = int(x / 3) * 3
    r_end = r_start + 3

    c_start = int(y / 3) * 3
    c_end = c_start + 3

    for i in range(r_start, r_end):
        for j in range(c_start, c_end):
            if i != x and j != y and sudoku[i][j] == key:
                return False
    return True

def fill_matrix(k, keys, r, rows):
    for c in graph[keys[k]][rows[r]]:
        if sudoku[rows[r]][c] > 0:
            continue
        sudoku[rows[r]][c] = keys[k]
        if is_safe(rows[r], c):
            if r < len(rows) - 1:
                if fill_matrix(k, keys, r + 1, rows):
                    return True
                else:
                    sudoku[rows[r]][c] = 0
                    continue
            else:
                if k < len(keys) - 1:
                    if fill_matrix(k + 1, keys, 0, list(graph[keys[k + 1]].keys())):
                        return True
                    else:
                        sudoku[rows[r]][c] = 0
                        continue
                return True
        sudoku[rows[r]][c] = 0
    return False


# Fill the pos and rem dictionary. It will be used to build graph
def build_pos_and_rem():
    for i in range(0, 9):
        for j in range(0, 9):
            if sudoku[i][j] > 0:
                if sudoku[i][j] not in pos:
                    pos[sudoku[i][j]] = []
                pos[sudoku[i][j]].append([i, j])
                if sudoku[i][j] not in rem:
                    rem[sudoku[i][j]] = 9
                rem[sudoku[i][j]] -= 1

    # Fill the elements not present in input matrix. Example: 1 is missing in input matrix
    for i in range(1, 10):
        if i not in pos:
            pos[i] = []
        if i not in rem:
            rem[i] = 9

# Build the graph
def build_graph():
    for k, v in pos.items():
        if k not in graph:
            graph[k] = {}

        row = list(range(0, 9))
        col = list(range(0, 9))

        for cord in v:
            row.remove(cord[0])
            col.remove(cord[1])

        if len(row) == 0 or len(col) == 0:
            continue

        for r in row:
            for c in col:
                if sudoku[r][c] == 0:
                    if r not in graph[k]:
                        graph[k][r] = []
                    graph[k][r].append(c)


build_pos_and_rem()

# Sort the rem map in order to start with smaller number of elements to be filled first. Optimization for pruning
rem = {k: v for k, v in sorted(rem.items(), key=lambda item: item[1])}

build_graph()

key_s = list(rem.keys())
# Util called to fill the matrix
fill_matrix(0, key_s, 0, list(graph[key_s[0]].keys()))

print("\nSolved Matrix:")
printMatrix()
```
